## Part III: Functions

### 1. Write a function to convert temperature from kelvin to celsius and celsius to kevin

```python
def kelvin_to_celsius(k):
    return k - 273.15

def celsius_to_kelvin(c):
    return c + 273.15
```

## 2. Write a function to convert temperature to fahrenheit

```python
def to_fahrenheit(temp, unit="celsius"):
    """ Convert temperature to Fahrenheit """
    if unit.lower() == "celsius":
        return (temp * 9/5) + 32
    elif unit.lower() == "kelvin":
        c = kelvin_to_celsius(temp)
        return (c * 9/5) + 32
```

### 4) Now write a function that converts from farenheit

```python
def from_fahrenheit(f, output_unit="celsius"):
    """ Convert temperature from Fahrenheit """
    c = (f - 32) * 5/9
    
    if output_unit.lower() == "celsius":
        return c
    elif output_unit.lower() == "kelvin":
        return celsius_to_kelvin(c) 
```

```python
print(from_fahrenheit(32, "kelvin"))
```

### 5) Write a function that takes two arguments (feet and inches) and returns height in meters

```python
def imperial_to_meters(feet, inches):
    """ Convert feet and inches to meters """
    total_inches = feet * 12 + inches
    return total_inches * 0.0254
```

```python
#Verifying
print(imperial_to_meters(6, 0))
```

```python
#Verifying
print(imperial_to_meters(5, 5))
```

### 6. Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

```python
def meters_to_imperial(meters):
    """ Converting meters to ft and inches """
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches % 12
    return feet, inches
```
